### Mappatura ATECO Caricata

**Struttura del dizionario:**
- **Colonna 1**: Codice ATECO (tutti i livelli gerarchici)
- **Colonna 2**: Descrizione testuale del settore

**Caratteristiche rilevate:**
- 3.157 codici totali (da lettere singole a codici a 6 cifre)
- Alcuni codici contengono **punti** (es. "01.11") che vanno rimossi per il matching
- Nomi colonne con caratteri speciali (`\n`) che vanno normalizzati

**Distribuzione per livello:**
- Livello 1 (Lettere): 21 codici (es. "A", "C", "G")
- Livello 2 (2 cifre): 88 codici (es. "01", "62")
- Livelli 3-6: 3.048 codici dettagliati

 

```python
# CELLA 7: Pulizia Codici ATECO nel Mapping
print("\n[FASE 3] Pulizia e normalizzazione codici ATECO")

# Rinomina colonne per facilità di utilizzo
ateco_mapping.columns = ['Codice_ATECO', 'Descrizione_ATECO']
print("✓ Colonne rinominate")

# Mostra esempi prima della pulizia
print(f"\nEsempi PRIMA della pulizia: {ateco_mapping['Codice_ATECO'].head(10).tolist()}")

# Rimuovi punti dai codici ATECO
ateco_mapping['Codice_ATECO'] = ateco_mapping['Codice_ATECO'].astype(str).str.replace('.', '', regex=False)
print("✓ Punti rimossi dai codici ATECO")

# Rimuovi spazi bianchi
ateco_mapping['Codice_ATECO'] = ateco_mapping['Codice_ATECO'].str.strip()
ateco_mapping['Descrizione_ATECO'] = ateco_mapping['Descrizione_ATECO'].str.strip()

print(f"Esempi DOPO la pulizia: {ateco_mapping['Codice_ATECO'].head(10).tolist()}")

# Verifica lunghezze codici
print(f"\nDistribuzione lunghezze codici ATECO:")
print(ateco_mapping['Codice_ATECO'].str.len().value_counts().sort_index())

print("\n--- Prime 10 righe dopo pulizia ---")
print(ateco_mapping.head(10))
```

## ---------------------------------------
### Codici ATECO Puliti e Normalizzati

**Operazioni eseguite:**
1. **Rinominate colonne**: da nomi complessi a `Codice_ATECO` e `Descrizione_ATECO`
2. **Rimossi punti**: da "01.11" a "0111" per matching corretto
3. **Eliminati spazi bianchi** extra

**Esempio trasformazione:**
- Prima: `"01.11"` → Dopo: `"0111"`
- Prima: `"62.01"` → Dopo: `"6201"`

**Distribuzione lunghezze codici:**
- 1 carattere: 21 codici (macro-settori A-U)
- 2 caratteri: 88 codici (divisioni)
- 3 caratteri: 272 codici (gruppi)
- 4 caratteri: 615 codici (classi)
- 5 caratteri: 920 codici (categorie)
- 6 caratteri: 1.241 codici (sottocategorie)

Il dizionario è  **pronto per il join** con i dataset 2019 e 2025.

```python
# CELLA 8: Normalizzazione Codici nei Dataset Principali
print("\n[Normalizzazione codici nei dataset 2019 e 2025]")

# Lista colonne con codici ATECO
codice_cols = ['Settore', 'Divisione', 'Classe', 'Sottocategoria']

for col in codice_cols:
    if col in df_2019.columns:
        # Rimuovi spazi bianchi
        df_2019[col] = df_2019[col].astype(str).str.strip()
        df_2025[col] = df_2025[col].astype(str).str.strip()
        
        print(f"✓ Pulita colonna: {col}")

print("✓ Normalizzazione codici completata")
```

## -------------------------------------------------------------
### Normalizzazione Codici nei Dataset Principali

**Operazioni di pulizia completate:**
- Rimossi spazi bianchi da colonne `Settore`, `Divisione`, `Classe`, `Sottocategoria`
- Codici uniformati e pronti per il matching con il dizionario ATECO

I codici nei dataset 2019 e 2025 sono ora **consistenti** con quelli del dizionario.

**Prossimo step**: Convertire Addetti e Localizzazioni da stringhe a numeri.

```python
# CELLA 9: Conversione Valori Numerici
print("\n[FASE 4] Conversione valori numerici")

# Converti Addetti e Localizzazioni Attive in numeri
df_2019['Addetti'] = pd.to_numeric(df_2019['Addetti'], errors='coerce')
df_2019['Localizzazioni Attive'] = pd.to_numeric(df_2019['Localizzazioni Attive'], errors='coerce')

df_2025['Addetti'] = pd.to_numeric(df_2025['Addetti'], errors='coerce')
df_2025['Localizzazioni Attive'] = pd.to_numeric(df_2025['Localizzazioni Attive'], errors='coerce')

# Verifica valori mancanti
print(f"Valori mancanti 2019 - Addetti: {df_2019['Addetti'].isna().sum()}, Localizzazioni: {df_2019['Localizzazioni Attive'].isna().sum()}")
print(f"Valori mancanti 2025 - Addetti: {df_2025['Addetti'].isna().sum()}, Localizzazioni: {df_2025['Localizzazioni Attive'].isna().sum()}")

# Rimuovi righe con valori mancanti
df_2019 = df_2019.dropna(subset=['Addetti', 'Localizzazioni Attive'])
df_2025 = df_2025.dropna(subset=['Addetti', 'Localizzazioni Attive'])

print(f"✓ Dataset 2019 dopo pulizia: {len(df_2019):,} righe")
print(f"✓ Dataset 2025 dopo pulizia: {len(df_2025):,} righe")
```

### Valori Numerici Convertiti

**Conversione completata:**
- Colonne `Addetti` e `Localizzazioni Attive` trasformate da stringhe a numeri
- **Nessun valore mancante** rilevato nei campi numerici
- Dataset puliti: 118.814 righe (2019) e 114.591 righe (2025)

I dataset sono ora **pronti per le analisi statistiche**.

```python
# CELLA 10: Statistiche Descrittive Nazionali
print("\n[FASE 5] Statistiche descrittive ITALIA")

print("\n" + "="*80)
print("ITALIA 2019 Q4")
print("="*80)
totale_addetti_2019 = df_2019['Addetti'].sum()
totale_loc_2019 = df_2019['Localizzazioni Attive'].sum()
dim_media_2019 = totale_addetti_2019 / totale_loc_2019

print(f"Totale Addetti: {totale_addetti_2019:,.0f}")
print(f"Totale Unità Locali: {totale_loc_2019:,.0f}")
print(f"Dimensione Media Impresa: {dim_media_2019:.2f} addetti/sede")

print("\n" + "="*80)
print("ITALIA 2025 Q3")
print("="*80)
totale_addetti_2025 = df_2025['Addetti'].sum()
totale_loc_2025 = df_2025['Localizzazioni Attive'].sum()
dim_media_2025 = totale_addetti_2025 / totale_loc_2025

print(f"Totale Addetti: {totale_addetti_2025:,.0f}")
print(f"Totale Unità Locali: {totale_loc_2025:,.0f}")
print(f"Dimensione Media Impresa: {dim_media_2025:.2f} addetti/sede")
```

### Statistiche Descrittive Italia

**Italia 2019 Q4 (Pre-Pandemia):**
- Addetti totali: **18.061.839**
- Unità locali: **6.342.757**
- Dimensione media impresa: **2.85 addetti/sede**

**Italia 2025 Q3 (Post-Pandemia):**
- Addetti totali: **19.839.628**
- Unità locali: **6.420.818**
- Dimensione media impresa: **3.09 addetti/sede**

**Interpretazione iniziale:**
- Il Paese ha recuperato e superato i livelli occupazionali pre-pandemia
- L'aumento di unità locali è modesto rispetto alla crescita degli addetti
- Le imprese tendono a diventare più grandi (consolidamento del mercato)

```python
# CELLA 11: Calcolo Variazioni Aggregate
print("\n" + "="*80)
print("VARIAZIONE 2019-2025")
print("="*80)

var_addetti = totale_addetti_2025 - totale_addetti_2019
var_perc_addetti = (var_addetti / totale_addetti_2019) * 100

var_localizzazioni = totale_loc_2025 - totale_loc_2019
var_perc_loc = (var_localizzazioni / totale_loc_2019) * 100

var_dim_media = dim_media_2025 - dim_media_2019
var_perc_dim_media = (var_dim_media / dim_media_2019) * 100

print(f"Variazione Addetti: {var_addetti:+,.0f} ({var_perc_addetti:+.2f}%)")
print(f"Variazione Unità Locali: {var_localizzazioni:+,.0f} ({var_perc_loc:+.2f}%)")
print(f"Variazione Dimensione Media: {var_dim_media:+.2f} addetti/sede ({var_perc_dim_media:+.2f}%)")

print("\n" + "="*80)
print(" CARICAMENTO COMPLETATO - Pronto per le analisi")
print("="*80)
```

### Variazione Aggregata 2019-2025

**Conclusione fase setup:**
I dati sono stati caricati, puliti e validati. Procediamo ora con le analisi settoriali e territoriali dettagliate.

```python
# CELLA 12: Aggregazione per Settore (Livello Divisione)
print("\n" + "="*80)
print("ANALISI 1: CRESCITA PER SETTORE (LIVELLO DIVISIONE)")
print("="*80)

# Aggrega dati per Divisione (2 cifre - es. "01", "62")
settori_2019 = df_2019.groupby('Divisione').agg({
    'Addetti': 'sum',
    'Localizzazioni Attive': 'sum'
}).reset_index()

settori_2025 = df_2025.groupby('Divisione').agg({
    'Addetti': 'sum',
    'Localizzazioni Attive': 'sum'
}).reset_index()

# Rinomina colonne per chiarezza
settori_2019.columns = ['Divisione', 'Addetti_2019', 'Localizzazioni_2019']
settori_2025.columns = ['Divisione', 'Addetti_2025', 'Localizzazioni_2025']

# Merge dei due periodi
confronto_settori = settori_2019.merge(settori_2025, on='Divisione', how='outer')

# Riempi valori mancanti con 0
confronto_settori = confronto_settori.fillna(0)

# Calcola variazioni
confronto_settori['Var_Addetti'] = confronto_settori['Addetti_2025'] - confronto_settori['Addetti_2019']
confronto_settori['Var_Perc_Addetti'] = (confronto_settori['Var_Addetti'] / confronto_settori['Addetti_2019'].replace(0, 1)) * 100
confronto_settori['Var_Localizzazioni'] = confronto_settori['Localizzazioni_2025'] - confronto_settori['Localizzazioni_2019']

# Aggiungi descrizioni ATECO
confronto_settori = confronto_settori.merge(
    ateco_mapping[ateco_mapping['Codice_ATECO'].str.len() == 2][['Codice_ATECO', 'Descrizione_ATECO']],
    left_on='Divisione',
    right_on='Codice_ATECO',
    how='left'
)

print(f"✓ Analizzati {len(confronto_settori)} settori (divisioni ATECO)")
print("\n--- Prime 10 righe confronto settori ---")
print(confronto_settori[['Divisione', 'Descrizione_ATECO', 'Addetti_2019', 'Addetti_2025', 'Var_Addetti', 'Var_Perc_Addetti']].head(10))
```

### Aggregazione per Settore (Livello Divisione)

**Metodologia:**
Abbiamo aggregato i dati a livello di **Divisione ATECO** (2 cifre), che rappresenta un buon compromesso tra:
- Dettaglio settoriale (88 divisioni vs 21 macro-settori)
- Significatività statistica (evita micro-settori con pochi addetti)

**Risultato:**
- Analizzati **89 settori** distinti
- Ogni settore ha ora dati aggregati per addetti e localizzazioni nei due periodi
- Calcolate variazioni assolute e percentuali

**Prime osservazioni:**
- Alcuni settori mostrano crescite molto elevate (>100%)
- Altri settori sono in decrescita strutturale
- Le descrizioni ATECO sono state aggiunte per leggibilità

 

```python
# CELLA 13: Top 10 Settori in Crescita Assoluta
print("\n" + "="*80)
print("TOP 10 SETTORI IN CRESCITA ASSOLUTA")
print("="*80)

top_10_crescita = confronto_settori.nlargest(10, 'Var_Addetti')[
    ['Divisione', 'Descrizione_ATECO', 'Addetti_2019', 'Addetti_2025', 'Var_Addetti', 'Var_Perc_Addetti']
]

print(top_10_crescita.to_string(index=False))

# Grafico Top 10 Crescita
plt.figure(figsize=(14, 8))
top_10_plot = confronto_settori.nlargest(10, 'Var_Addetti')

# Crea etichette combinate (codice + descrizione troncata)
labels = [f"{row['Divisione']}: {row['Descrizione_ATECO'][:40]}..." 
          if pd.notna(row['Descrizione_ATECO']) and len(row['Descrizione_ATECO']) > 40
          else f"{row['Divisione']}: {row['Descrizione_ATECO']}"
          for _, row in top_10_plot.iterrows()]

plt.barh(range(len(top_10_plot)), top_10_plot['Var_Addetti'], color='green', alpha=0.7)
plt.yticks(range(len(top_10_plot)), labels)
plt.xlabel('Variazione Addetti (2019-2025)', fontsize=12)
plt.title('Top 10 Settori per Crescita Occupazionale\n(2019 Q4 - 2025 Q3)', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)

# Aggiungi valori sulle barre
for i, (_, row) in enumerate(top_10_plot.iterrows()):
    plt.text(row['Var_Addetti'], i, f" +{row['Var_Addetti']:,.0f} ({row['Var_Perc_Addetti']:+.1f}%)", 
             va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# CELLA 14: Top 10 Settori in Decrescita
print("\n" + "="*80)
print("TOP 10 SETTORI IN DECRESCITA")
print("="*80)

top_10_decrescita = confronto_settori.nsmallest(10, 'Var_Addetti')[
    ['Divisione', 'Descrizione_ATECO', 'Addetti_2019', 'Addetti_2025', 'Var_Addetti', 'Var_Perc_Addetti']
]

print(top_10_decrescita.to_string(index=False))

# Grafico Top 10 Decrescita
plt.figure(figsize=(14, 8))
top_10_dec_plot = confronto_settori.nsmallest(10, 'Var_Addetti')

labels = [f"{row['Divisione']}: {row['Descrizione_ATECO'][:40]}..." 
          if pd.notna(row['Descrizione_ATECO']) and len(row['Descrizione_ATECO']) > 40
          else f"{row['Divisione']}: {row['Descrizione_ATECO']}"
          for _, row in top_10_dec_plot.iterrows()]

plt.barh(range(len(top_10_dec_plot)), top_10_dec_plot['Var_Addetti'], color='red', alpha=0.7)
plt.yticks(range(len(top_10_dec_plot)), labels)
plt.xlabel('Variazione Addetti (2019-2025)', fontsize=12)
plt.title('Top 10 Settori in Decrescita Occupazionale\n(2019 Q4 - 2025 Q3)', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)

# Aggiungi valori sulle barre
for i, (_, row) in enumerate(top_10_dec_plot.iterrows()):
    plt.text(row['Var_Addetti'], i, f" {row['Var_Addetti']:,.0f} ({row['Var_Perc_Addetti']:+.1f}%)", 
             va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# CELLA 23: Verifica Qualità Dati e Codici Settore
print("\n" + "="*80)
print("ANALISI QUALITÀ DATI: DISTRIBUZIONE CODICI SETTORE")
print("="*80)

print("\n--- Distribuzione Settori 2019 ---")
settori_dist_2019 = df_2019['Settore'].value_counts().sort_index()
print(f"Totale settori diversi: {len(settori_dist_2019)}")
print(settori_dist_2019)

print("\n--- Distribuzione Settori 2025 ---")
settori_dist_2025 = df_2025['Settore'].value_counts().sort_index()
print(f"Totale settori diversi: {len(settori_dist_2025)}")
print(settori_dist_2025)

# Confronta presenza/assenza settori
print("\n--- Analisi Comparativa ---")
set_2019 = set(settori_dist_2019.index)
set_2025 = set(settori_dist_2025.index)

settori_scomparsi = set_2019 - set_2025
settori_nuovi = set_2025 - set_2019

if settori_scomparsi:
    print(f"\n Settori presenti nel 2019 ma ASSENTI nel 2025: {sorted(settori_scomparsi)}")
    for s in sorted(settori_scomparsi):
        addetti = df_2019[df_2019['Settore']==s]['Addetti'].sum()
        desc = ateco_mapping[ateco_mapping['Codice_ATECO']==s]['Descrizione_ATECO'].values
        desc_str = desc[0] if len(desc) > 0 else 'N/A'
        print(f"  {s}: {addetti:>12,.0f} addetti - {desc_str}")

if settori_nuovi:
    print(f"\n Settori NUOVI nel 2025 (assenti nel 2019): {sorted(settori_nuovi)}")
    for s in sorted(settori_nuovi):
        addetti = df_2025[df_2025['Settore']==s]['Addetti'].sum()
        desc = ateco_mapping[ateco_mapping['Codice_ATECO']==s]['Descrizione_ATECO'].values
        desc_str = desc[0] if len(desc) > 0 else 'N/A'
        print(f"  {s}: {addetti:>12,.0f} addetti - {desc_str}")
```

```python
# CELLA 25: Analisi Geografica Dettagliata - Nord vs Centro vs Sud
print("\n" + "="*80)
print("ANALISI GEOGRAFICA: NORD vs CENTRO vs SUD")
print("="*80)

# Classificazione regioni
nord = ['PIEMONTE', 'VALLE D\'AOSTA', 'LOMBARDIA', 'TRENTINO - ALTO ADIGE', 
        'VENETO', 'FRIULI-VENEZIA GIULIA', 'LIGURIA', 'EMILIA ROMAGNA']
centro = ['TOSCANA', 'UMBRIA', 'MARCHE', 'LAZIO']
sud = ['ABRUZZO', 'MOLISE', 'CAMPANIA', 'PUGLIA', 'BASILICATA', 
       'CALABRIA', 'SICILIA', 'SARDEGNA']

confronto_regioni['Area'] = confronto_regioni['Regione'].apply(
    lambda x: 'Nord' if x in nord else ('Centro' if x in centro else 'Sud')
)

# Aggrega per area
aree_stats = confronto_regioni.groupby('Area').agg({
    'Addetti_2019': 'sum',
    'Addetti_2025': 'sum',
    'Localizzazioni_2019': 'sum',
    'Localizzazioni_2025': 'sum',
    'Var_Addetti': 'sum'
}).reset_index()

aree_stats['Var_Perc_Addetti'] = (aree_stats['Var_Addetti'] / aree_stats['Addetti_2019']) * 100
aree_stats['Dim_Media_2019'] = aree_stats['Addetti_2019'] / aree_stats['Localizzazioni_2019']
aree_stats['Dim_Media_2025'] = aree_stats['Addetti_2025'] / aree_stats['Localizzazioni_2025']
aree_stats['Peso_2019'] = (aree_stats['Addetti_2019'] / aree_stats['Addetti_2019'].sum()) * 100
aree_stats['Peso_2025'] = (aree_stats['Addetti_2025'] / aree_stats['Addetti_2025'].sum()) * 100

print("\n--- Statistiche per Macro-Area ---")
print(aree_stats.to_string(index=False))

# Grafico comparativo aree
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# 1. Occupazione per area
x = range(len(aree_stats))
width = 0.35
ax1.bar([i - width/2 for i in x], aree_stats['Addetti_2019'], width, 
        label='2019 Q4', color='steelblue', alpha=0.7)
ax1.bar([i + width/2 for i in x], aree_stats['Addetti_2025'], width, 
        label='2025 Q3', color='orange', alpha=0.7)
ax1.set_xticks(x)
ax1.set_xticklabels(aree_stats['Area'])
ax1.set_ylabel('Numero Addetti (milioni)', fontsize=11)
ax1.set_title('Occupazione per Macro-Area', fontsize=12, fontweight='bold')
ax1.legend()
ax1.grid(axis='y', alpha=0.3)
# Converti in milioni
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))

# 2. Crescita percentuale
colors_area = ['green' if x > aree_stats['Var_Perc_Addetti'].mean() else 'orange' 
               for x in aree_stats['Var_Perc_Addetti']]
ax2.bar(aree_stats['Area'], aree_stats['Var_Perc_Addetti'], color=colors_area, alpha=0.7)
ax2.axhline(y=aree_stats['Var_Perc_Addetti'].mean(), color='red', 
            linestyle='--', label=f'Media Italia: {aree_stats["Var_Perc_Addetti"].mean():.1f}%')
ax2.set_ylabel('Variazione % Addetti', fontsize=11)
ax2.set_title('Crescita Occupazionale per Area', fontsize=12, fontweight='bold')
ax2.legend()
ax2.grid(axis='y', alpha=0.3)
for i, (_, row) in enumerate(aree_stats.iterrows()):
    ax2.text(i, row['Var_Perc_Addetti'], f"{row['Var_Perc_Addetti']:.1f}%", 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# 3. Dimensione media imprese
ax3.bar([i - width/2 for i in x], aree_stats['Dim_Media_2019'], width, 
        label='2019 Q4', color='steelblue', alpha=0.7)
ax3.bar([i + width/2 for i in x], aree_stats['Dim_Media_2025'], width, 
        label='2025 Q3', color='orange', alpha=0.7)
ax3.set_xticks(x)
ax3.set_xticklabels(aree_stats['Area'])
ax3.set_ylabel('Addetti per Sede', fontsize=11)
ax3.set_title('Dimensione Media Imprese per Area', fontsize=12, fontweight='bold')
ax3.legend()
ax3.grid(axis='y', alpha=0.3)

# 4. Peso sul totale nazionale
ax4.bar([i - width/2 for i in x], aree_stats['Peso_2019'], width, 
        label='2019 Q4', color='steelblue', alpha=0.7)
ax4.bar([i + width/2 for i in x], aree_stats['Peso_2025'], width, 
        label='2025 Q3', color='orange', alpha=0.7)
ax4.set_xticks(x)
ax4.set_xticklabels(aree_stats['Area'])
ax4.set_ylabel('% sul Totale Italia', fontsize=11)
ax4.set_title('Peso Economico per Area', fontsize=12, fontweight='bold')
ax4.legend()
ax4.grid(axis='y', alpha=0.3)
for i, (_, row) in enumerate(aree_stats.iterrows()):
    ax4.text(i - width/2, row['Peso_2019'], f"{row['Peso_2019']:.1f}%", 
             ha='center', va='bottom', fontsize=8)
    ax4.text(i + width/2, row['Peso_2025'], f"{row['Peso_2025']:.1f}%", 
             ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# CELLA 25.1: Box Plot Distribuzione Crescita per Area


print("\n" + "="*80)
print("BOX PLOT: DISTRIBUZIONE CRESCITA REGIONALE PER AREA")
print("="*80)

# Prepara dati per box plot
box_data = []
for area in ['Nord', 'Centro', 'Sud']:
    regioni_area = confronto_regioni[confronto_regioni['Area'] == area]['Var_Perc_Addetti'].values
    box_data.append(regioni_area)

# Crea box plot
fig, ax = plt.subplots(figsize=(14, 8))

bp = ax.boxplot(box_data, 
                labels=['Nord', 'Centro', 'Sud'],
                patch_artist=True,
                notch=True,
                showmeans=True,
                meanprops=dict(marker='D', markerfacecolor='red', markersize=8))

# Colora i box
colors = ['steelblue', 'orange', 'green']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)

# Aggiungi punti individuali (regioni)
for i, area in enumerate(['Nord', 'Centro', 'Sud'], 1):
    regioni_area = confronto_regioni[confronto_regioni['Area'] == area]
    y_values = regioni_area['Var_Perc_Addetti'].values
    x_values = np.random.normal(i, 0.04, size=len(y_values))  # Jitter per visibilità
    ax.scatter(x_values, y_values, alpha=0.5, s=100, color='black', edgecolors='white', linewidth=1)
    
    # Etichetta outlier
    for idx, row in regioni_area.iterrows():
        if row['Var_Perc_Addetti'] > regioni_area['Var_Perc_Addetti'].quantile(0.75) + 1.5 * regioni_area['Var_Perc_Addetti'].std() or \
           row['Var_Perc_Addetti'] < regioni_area['Var_Perc_Addetti'].quantile(0.25) - 1.5 * regioni_area['Var_Perc_Addetti'].std():
            ax.annotate(row['Regione'], 
                       xy=(i, row['Var_Perc_Addetti']),
                       xytext=(10, 0),
                       textcoords='offset points',
                       fontsize=8,
                       bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.5))

ax.set_ylabel('Variazione % Addetti (2019-2025)', fontsize=12, fontweight='bold')
ax.set_title('Distribuzione della Crescita Occupazionale per Macro-Area\n(Box = Q1-Q3, Linea = Mediana, Rombo = Media)', 
             fontsize=14, fontweight='bold', pad=20)
ax.grid(axis='y', alpha=0.3)
ax.axhline(y=var_perc_addetti, color='red', linestyle='--', linewidth=2, label=f'Media Italia: {var_perc_addetti:.1f}%')
ax.legend(fontsize=10)

plt.tight_layout()
plt.show()

# Statistiche descrittive
print("\n--- Statistiche Descrittive per Area ---")
for area in ['Nord', 'Centro', 'Sud']:
    regioni_area = confronto_regioni[confronto_regioni['Area'] == area]['Var_Perc_Addetti']
    print(f"\n{area}:")
    print(f"  Media: {regioni_area.mean():.2f}%")
    print(f"  Mediana: {regioni_area.median():.2f}%")
    print(f"  Dev. Std: {regioni_area.std():.2f}%")
    print(f"  Min: {regioni_area.min():.2f}% | Max: {regioni_area.max():.2f}%")
    print(f"  Range interquartile: {regioni_area.quantile(0.75) - regioni_area.quantile(0.25):.2f}%")

print("\n--- Interpretazione ---")
print("Box più ampio = Maggiore eterogeneità interna")
print("Punti fuori dai baffi = Regioni outlier")
print(f"Sud: Mediana più alta ({confronto_regioni[confronto_regioni['Area']=='Sud']['Var_Perc_Addetti'].median():.1f}%) ma anche più variabilità")
```

**Centro:**

- Range: 3-15%

**Sud:**
